**Exploratory Data Analysis**

#### Distribution of tv show and movie that released between 2016 to 2021 in Indonesia, Malaysia, Singapore

```python
df_copy = df[['type', 'title', 'country', 'release_year']].copy()
```

```python
df_id = df_copy[df_copy['country'] == "Indonesia"]
df_id = df_id[(df_id['release_year'] <= 2021) & (df_id['release_year'] >= 2016)]
df_id.head(10)
```

```python
df_my = df_copy[df_copy['country'] == "Malaysia"]
df_my = df_my[(df_my['release_year'] <= 2021) & (df_my['release_year'] >= 2016)]
df_my.head(10)
```

```python
df_sg = df_copy[df_copy['country'] == "Singapore"]
df_sg = df_sg[(df_sg['release_year'] <= 2021) & (df_sg['release_year'] >= 2016)]
df_sg.head(10)
```
